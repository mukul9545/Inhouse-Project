```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data
df = pd.read_csv('_processed_tweets.csv')

# Step 2: Parse dates
df['createdAt'] = pd.to_datetime(df['createdAt'])

# Step 3: Aggregate sentiment over time
# Resample by day and calculate the mean sentiment
sentiment_over_time = df.resample('30D', on='createdAt')['polarity'].mean()

# Step 4: Plot the sentiment over time
plt.figure(figsize=(12, 6))
plt.plot(sentiment_over_time, marker='o', linestyle='-', color='b')
plt.title('Sentiment Over Time (30-Day Average)')
plt.xlabel('Date')
plt.ylabel('Average Sentiment')
plt.grid(True)

```
