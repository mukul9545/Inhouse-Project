```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from matplotlib import style
from collections import Counter

style.use('ggplot')
df = pd.read_csv("swiggyinstamart_processed_tweets.csv")

# Separate tweets by sentiment
pos_tweets = df[df.sentiment == 'Positive']
neg_tweets = df[df.sentiment == 'Negative']
neutral_tweets = df[df.sentiment == 'Neutral']

# Function to generate word cloud and print top 20 frequent terms
def generate_wordcloud_and_print_top50(text, title):
    # Generate word cloud
    wordcloud = WordCloud(max_words=500, width=1600, height=800).generate(text)
    
    # Plot word cloud
    plt.figure(figsize=(20,15), facecolor='None')
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title, fontsize=19)
    plt.show()
    
    # Print top 20 frequent terms
    words = text.split()
    word_freq = Counter(words)
    top50_words = word_freq.most_common(50)
    print("Top 50 most frequent terms in", title)
    for word, freq in top50_words:
        print(word, ":", freq)

# Generate word cloud and print top 20 frequent terms for positive tweets
pos_text = ' '.join([word for word in pos_tweets['processed_text']])
generate_wordcloud_and_print_top50(pos_text, 'Positive Tweets')

# Generate word cloud and print top 20 frequent terms for negative tweets
neg_text = ' '.join([word for word in neg_tweets['processed_text']])
generate_wordcloud_and_print_top50(neg_text, 'Negative Tweets')

# Generate word cloud and print top 20 frequent terms for neutral tweets
neutral_text = ' '.join([word for word in neutral_tweets['processed_text']])
generate_wordcloud_and_print_top50(neutral_text, 'Neutral Tweets')
```
